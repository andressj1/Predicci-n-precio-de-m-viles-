```python
###-----librerias-----###
import pandas as pd 
import plotly.express as px    ##Crear gráficos interactivos
import seaborn as sns # Matriz de correlacion
import matplotlib.pyplot as plt #Graficos
```

```python
###----Lectura de datos----###
df = pd.read_csv("https://raw.githubusercontent.com/andressj1/Predicci-n-precio-de-m-viles-/refs/heads/main/Datos/train.csv", sep=",")
```

# Analisis Univariado 

Variables Categoricas

```python
# Lista de las variables categoricas
variables_categoricas = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi', 'price_range']
```

```python
# Iterar sobre cada variable
for var in variables_categoricas:
    # Obtener los conteos de los valores de la variable actual
    attrition_counts = df[var].value_counts()
    
    # Crear el gráfico de torta para la variable actual
    fig = px.pie(values=attrition_counts, names=attrition_counts.index, title=f'Gráfico de torta para {var}')
    
    # Mostrar el gráfico
    fig.show()
```

Variables Numericas

```python
variables_numericas  = [col for col in df.columns if col not in variables_categoricas]
print(variables_numericas)
```

```python
for var in variables_numericas:
    # Crear el histograma
    fig = px.histogram(df, x=var, title=f'Distribución de {var}')
    
    # Mostrar el gráfico
    fig.show()
```

# Matriz de correlación

```python
correlation_matrix = df[variables_numericas].corr()

plt.figure(figsize=(14, 14))
sns.set(font_scale=1.2)
sns.set_style("whitegrid")

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0,
            linewidths=.5, square=True, cbar_kws={"shrink": .5})
plt.title('Matriz de Correlación')
plt.show()
```
